# If-sats

I programmering så behöver man ofta använda *villkor*. Ett exempel på villkor kan vara

    Om temperaturen är över 25 grader, kör fläkten
    Om temperaturen är över 24 grader, stanna fläkten

Detta kan man i programmering styra med hjälp av en *if-sats*

Se exemplet nedan:

```python
temperatur = 27

if temperatur > 26:
    print("kör fläkt, vroooooooooooooom")
elif temperatur < 25:
    print("stanna fläkt")
```
