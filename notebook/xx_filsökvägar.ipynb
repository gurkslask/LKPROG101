# Filsökvägar

Filsökvägar är något som är viktigt inom programmering. Alla våra filer finns någonstans på hårddisken, och det är viktigt att **veta** var filerna finns och vart vi ska lägga olika filer. För att jobba med filsökvägar kan man använda **os** modulen i python. Nedan använder vi os-modulens path bibliotek för att ta reda på **sökvägen** för **mappen** som vi *står i*.

```python
import os

print(os.path.abspath("."))
```

--- 

Vi kan testa skapa en fil för att sedan visa vilken sökväg filen har

```python
with open("kagg.txt", "w") as f:
    f.write("Lars kagg")

absolut_sokvag = os.path.abspath("./kagg.txt")
print(absolut_sokvag)
print(os.path.basename(absolut_sokvag))
print(os.path.dirname(absolut_sokvag))
```

## Absolut och relativ sökväg

I kodexemplet ovan så vill jag att ni kollar extra noga på den här delen av koden:

    absolut_sokvag = os.path.abspath("./kagg.txt")

**./kagg.txt** är något som vi kallar en *relativ sökväg*. Med det menas att sökvägen är **relativ med var vi står just nu**. I koden så kallar vi på funktionen **abspath**, detta är en förkortning för **absolute path** som menas på svenska *en sökväg som funkar oavsett var vi befinner oss på hårddisken just nu*. I mitt fall blir sökvägarna såhär:


```python
print(f"Min absoluta sökväg är {absolut_sokvag}\nMin relativa sökväg är: ./kagg.txt")
```

--- 

Sökvägar är något som används **flitigt** och man blandar ofta relativa och absoluta sökvägar inom programmering. Kör man relativa sökvägar måste man vara **säker** på att man står på rätt ställe på hårddisken. Absoluta sökvägar används om man vet **exakt** var filen finns men man är osäker på vart programmet körs på hårddisken.

---

För att synliggöra ännu mer angående sökvägar, så kan man dela upp den i *mappnamn* och *filnamn*. Om vi kollar på den här sökvägen: **/home/alex/Projects/LKPROG101/notebook/kagg.txt** så delar jag upp den här under i *mappnamn* och *filnamn*

| /home/alex/Projects/LKPROG101/notebook/ | kagg.txt |
| --- | --- |
| mappnamn | filnamn |

Man kan enkelt säga att det **på slutet** är filnamnet, och det som är **innan** är *mappnamnet*

---

Om man kör relativ sökväg så **byter man ut** *mappnamnet* mot en relativ väg. Det finns olika relativa vägar, här kommer några vanliga:

 - **.** (Punkten står för mappen som vi är i för tillfället)
 - **..** (Punkt punkt står för mappen som är **ovanför** mappen som vi är i just nu)
 - **~** (~-tecknet kallas för **tilde**, och det står för vår användares **hemmapp** till exempel: */home/alex*
