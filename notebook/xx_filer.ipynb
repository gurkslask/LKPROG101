# Jobba med filer i python

Att jobba med filer i python är förhållandevis enkelt. För att öppna en fil är det *bara* att ange filnamn och instruktion om hur vi ska öppna filen. De val som finns när man ska öppna en fil är: 
 - r (read), read-only, alltså bara läsa data i filen. Om man öppnar den så här kan man **inte skriva** till filen
 - w (write), Om man öppnar den så här så skrivs **hela** filen över om man skriver till den
 - a (append), lägg till, om man öppnar den så här och skriver till den så hamnar det man skriver längst ner i filen

Nedan så kommer lite exempel på hur man kan jobba med filer. Det som är direkt här under är bara lite exempel data som vi kan ladda in i filerna.

```python
manader = [
    "Januari",
    "Februari",
    "Mars",
    "April",
]
dagar = [
    "Mandag",
    "Tisdag",
    "Onsdag",
    "Torsdag",
]
```

---

Nedan är ett exempel hur man kan öppna en fil för att skriva till filen. Tänk på att **första raden** är till för att **öppna** filen. **Andra raden** är där vi faktiskt **skriver** till filen.

```python
with open('test.txt', "w") as f:
    f.writelines(manader)
```

---

Nedan är ett exempel hur man kan öppna en fil för att **läsa** till filen. Tänk på att första raden är till för att öppna filen. Andra raden är där vi faktiskt **läser** filen.

```python
with open('test.txt', "r") as f:
    text = f.readlines()

print(text)
```

----
Här under testar vi att skriva till 
filen som jag öppnat med **"r"**. Om ni kör koden så kommer programmet bli argt på oss och säga att det inte går att skriva

```python
with open('test.txt', "r") as f:
    f.writelines(dagar)
    text = f.read()

print(text)
```

## Lägg till (append)

Nu ska vi testa att lägga till data till filen

---

Om vi testar skriva till filen igen så ska vi se vad som händer

I exemplet nedan skriver jag till filen och sen läser

```python
with open('test.txt', "r") as f:
    textUtanLines = f.read()

print(f"Detta är texten innan skrivning: {textUtanLines}\n")

with open('test.txt', "w") as f:
    f.writelines(dagar)

with open('test.txt', "r") as f:
    textUtanLines = f.read()

print(f"Detta är texten efter skrivning: {textUtanLines}\n")
```

--- 

Som ni ser skrivs all text över. Vill man lägga till text får man öppna filen med **a**. Titta på exemplet nedan där jag endast gjort en liten ändring

```python
with open('test.txt', "r") as f:
    textUtanLines = f.read()

print(f"Detta är texten innan skrivning: {textUtanLines}\n")

with open('test.txt', "a") as f: #HÄR HAR JAG GJORT EN ÄNDRING
    f.writelines(dagar)

with open('test.txt', "r") as f:
    textUtanLines = f.read()

print(f"Detta är texten efter skrivning: {textUtanLines}\n")
```
