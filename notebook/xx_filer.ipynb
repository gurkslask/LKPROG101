# Jobba med filer i python

Att jobba med filer i python är förhållandevis enkelt. För att öppna en fil är det *bara* att ange filnamn och instruktion om hur vi ska öppna filen. De val som finns när man ska öppna en fil är: 
 - r (read), read-only, alltså bara läsa data i filen. Om man öppnar den så här kan man **inte skriva** till filen
 - w (write), Om man öppnar den så här så skrivs **hela** filen över om man skriver till den
 - a (append), lägg till, om man öppnar den så här och skriver till den så hamnar det man skriver längst ner i filen

Nedan så kommer lite exempel på hur man kan jobba med filer. Det som är direkt här under är bara lite exempel data som vi kan ladda in i filerna.

```python
manader = [
    "Januari",
    "Februari",
    "Mars",
    "April",
]
dagar = [
    "Mandag",
    "Tisdag",
    "Onsdag",
    "Torsdag",
]
```

## Öppna en fil (write)

I exemplet öppnar vi en fil. Med hjälp av **"w"** så säger vi till programmet att vi ska öppna den i **skrivläge**. I filen så skriver vi datan från variabeln *manader*. När ni kört den här koden kan ni öppna filen som nu har skapats på er hårddisk och se att i filen finns det nu text.

```python
with open('test.txt', "w") as f:
    f.writelines(manader)
```

## Öppna en fil (read)

Nu ska vi läsa filen. Vi ändrar **"w"** till **"r"** för att läsa filen, vi kan nu alltså inte skriva till filen.

```python
with open('test.txt', "r") as f:
    text = f.readlines()

print(text)
```

----
Här under testar vi att skriva till 
filen som jag öppnat med **"r"**. Om ni kör koden så kommer programmet bli argt på oss och säga att det inte går att skriva

```python
with open('test.txt', "r") as f:
    f.writelines(dagar)
    text = f.read()

print(text)
```
