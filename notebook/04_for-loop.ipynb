# for-loop

for-loop är ett begrepp som används inom all programmering, oavsett vilket programmeringsspråk man använder så finns det alltid stöd för for-loopar. En for-loop är ett sätt för ett program att göra saker många gånger. **Det skulle tillexempel vara att göra en sak för varje sak i en lista**.

Titta på nedanstående exempel:


```python
betygAlternativ = [ "A", "B", "C", "D", "E", "F" ]

for betyg in betygAlternativ:
    print(betyg.lower())
```

----

I exemplet så går programmet igenom alla betyg i betygAlternativ-listan, och där skrivs betyg ut fast efter att de har ändrats till *gemener* istället för *VERSALER*.

--- 

Man kan också loopa och säga åt ett program hur många gånger den ska köra. Kolla på exemplet nedan:


```python
for nummer in range(10):
    print(f"Nu är vi på nummer: {nummer}")
```

Här körs loopen **10** gånger, men som ni märker börjar loopen på **0**. Det printas alltså ut 0-9.
